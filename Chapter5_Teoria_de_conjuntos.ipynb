 A Teoria de Conjuntos é a teoria matemática que se ocupa dum universo $U$ de
indivíduos a que chamamos conjuntos. Diz-se que os objecto $a$ e $A$
existem e são conjuntos, se pertencem ao domínio $U$. Vamos supor que se
verificam  certas relações fundamentais, da forma $a\in A$, entre
objectos $a$ e $A$ do domínio $U$. Se para dois objectos $a$ e $A$ vale a
relação $a\in A$, diremos que ''$a$ é um elemento de $A$'' ou que ''$A$ contém
$a$ como elemento''. Um conjunto designa-se geralmente por uma letra maiúscula, reservando-se as letras minúsculas para os seus elementos. A negação de $x\in A$ representa-se simbolicamente por $x\not \in A$.

### Descrição
Um conjunto pode ser descrito em **extensão** (quando o número dos seus elementos for finito e suficientemente pequeno) enumerando explicitamente todos os seus elementos e colocados entre chavetas e separados por vírgulas ou em **compreensão**, enunciando uma propriedade caracterizadora dos seus elementos (isto é, uma propriedade que os seus e só os seus elementos possuem).

#### Exemplo:
Conjunto das vogais $V$ é descrito em extensão $V=\{a,e,i,o,u\}$;


#### Exemplo:
Conjunto $L$ das palavras da lingua portuguesa escritas com 5 letras vogais, está descrito em compreensão;


#### Exemplo:
Conjunto dos números naturais pares pode ser descrito, em compreensão, como sendo definido por naturais $n$ para os quais existe um natural $q$ tal que $n=2q$.

#### Exemplo:
$1\in \{1,2,3\},\;\; 2 \in \{4,2,5\},\;\;7\not\in\{1,2,3\},\;\;\{1\}\not\in \{1,2,3\}$

#### Exemplo:
$1\in\{n:\; n \text{ é um natural ímpar}\}\;\;2\not\in\{n:\; n \text{ é um natural ímpar}\}$

### Axioma da extensão
Dois conjuntos $A$ e $B$ são iguais se e só se têm os mesmos
elementos. Neste caso escrevemos, $A=B$.

Simbolicamente temos $$A=B\Leftrightarrow \forall x(x\in A \leftrightarrow x\in B)$$
Dada uma expressão proposicional $P(x)$ (um predicado colectivizante), há um só conjunto
que tem por elementos os $x$ que satisfazem $P(x)$, tal conjunto
será denotado por $$\{x:P(x)\}.$$

#### Exemplo:
1. $\{1,2,3,4,5\} = \{5,4,1,2,3\}$ e $\{5,5,5,5\}=\{5\}$ porque os conjuntos envolvidos têm os mesmos elementos.
1. $\{1,2,3,4,5\} \neq \{5,4,2,3\}$ e $\{5,5,5,5\}\neq\{5,6\}$ porque os conjuntos envolvidos são definidos por elementos diferentes.


#### Proposição:
No domínio $U$ dos conjuntos temos a relação de igualdade de conjuntos satisfaz:
1. $\forall x(x=x)$ (reflexiva)
1. $\forall x\forall y(x=y\rightarrow y=x)$ (simétrica)
1. $\forall x\forall y\forall z(x=y \wedge y=z\rightarrow x=z)$

Quando não se quer ser tão formal, escrevemos de forma equivalente;

#### Proposição:
A relação de igualdade de conjuntos satisfaz, para todos os conjuntos $A$, $B$ e $C$:
1. $A=A$ (reflexiva)
1. $A=B \Rightarrow B=A$ (simétrica)
1. $A=B \wedge B=C\Rightarrow A=C$ (transitiva)

Enfraquecendo a relação de igualdade, diremos que $A$ é subconjunto de $B$ se todo o elemento de $A$ é um elemento de $B$, que podemos traduzir simbolicamente escrevendo:

$$A\subseteq B \Leftrightarrow \forall x (x\in A \rightarrow x\in B)$$

#### Exemplo
Se $A=\{\{1\},2,3\}$, então $\{1\}\in A$, $\{\{1\},2\}\subseteq A$, $2\in A$, $\{2,3\}\subseteq A$.

Com esta definição pode enunciar-se que para conjuntos $A$ e $B$:

"$A\subseteq  B$ e $B \subseteq A$ se e só se  $A = B$."

Propriedade que podemos demonstrar, recorrendo às propriedades da lógica proposicional:

$$A\subseteq  B \wedge B \subseteq A$$
$$\Leftrightarrow$$
$$\forall x (x\in A \rightarrow x\in B)\wedge\forall x (x\in B \rightarrow x\in A)$$
$$\Leftrightarrow$$
$$\forall x ((x\in A \rightarrow x\in B)\wedge (x\in B \rightarrow x\in A))$$
$$\Leftrightarrow$$
$$\forall x (x\in A \leftrightarrow x\in B)$$
$$\Leftrightarrow$$
$$A = B$$
   
De forma idêntica pode-se demonstrar a veracidade de cada uma das afirmações apresentadas na proposição abaixo:

#### Proposição
No domínio $U$ dos conjuntos valem para a relação de inclusão:

1. $\forall x(x\subseteq x)$ (reflexiva)
1. $\forall x\forall y(x\subseteq y\wedge y\subseteq x \rightarrow y=x)$ (anti-simétrica)
1. $\forall x\forall y\forall z(x\subseteq y \wedge y\subseteq z\rightarrow x\subseteq z)$(transitiva)

Para conjuntos $A$, $B$ e $C$ valem:

1. $A\subseteq A$ (reflexiva)
1. $(A\subseteq B\wedge B\subseteq A) \Rightarrow A=B$ (anti-simétrica)
1. $(A\subseteq B \wedge B\subseteq C) \Rightarrow A\subseteq C$ (transitiva)


## Axioma dos conjuntos elementares

Assumimos que no universo dos conjuntos existe um conjunto que não contém qualquer elemento, que denotamos por $\emptyset$, e que designamos de conjunto vazio. Se $a$ é um
conjunto, existe um conjunto que contém apenas o elemento $a$, que representamos em extensão por $\{a\}$, e que designamos de \textbf{conjunto singular}. Se $a$ e $b$ são conjuntos existe um conjunto que contém $a$ e $b$, e apenas $a$ e $b$ como elementos, que representamos em extensão por $\{a,b\}$, que designamos de par não ordenado.

## Esquema da separação

m predicado $P(x)$ diz-se colectivizante se define um conjunto. Se $P(x)$ é colectivizante, para cada conjunto
$A$ existe um conjunto cujos elementos são precisamente os elementos $a$ de $A$ para os quais $P(a)$ é verdadeira.


Isto é, para cada expressão proposicional colectivizante $P(x)$ e cada conjunto $A$, a fórmula $$x\in A \wedge P(x),$$ define um conjunto. Assegura-se assim que existe um conjunto definido pela expressão:$$\{x:x\in A \wedge P(x)\} \text{ ou }\{x\in A:P(x)\}$$

Assegurando-se assim a existência da intersecção de dois conjuntos $A$ e $B$,
  $$A\cap B=\{x:x\in A\wedge x\in B\}$$

#### Proposição:
Para conjuntos $A,B$ e $C$ valem
1. $A\cap A=A$ (idempotência)
- $A\cap \emptyset = \emptyset$
- $A\cap B=B \cap A$ (comutatividade)
- $(A\cap B)\cap C=A \cap( B\cap C)$ (transitividade)
- $(C\subseteq A\wedge C\subseteq B)\Rightarrow C\subseteq A\cap B$
- $A\subseteq B \Leftrightarrow A\cap B= A$

#### Definição: [Conjuntos Disjuntos]
Dois conjuntos $A$ e $B$ são disjuntos se e só se $A\cap B = \emptyset$, ou seja se não têm elementos em comum.

#### Exemplo:
Se $A_1=\{\{1,2\},\{3\}\}$, $A_2=\{\{1\},\{2,3\}\}$ e $A_3=\{\{1,2,3\}\}$, os conjuntos $A_1,A_2,A_3$ são disjuntos dois a dois, já que $A_1\cap A_2=\emptyset$, $A_1\cap A_3=\emptyset$ e $A_1\cap A_3=\emptyset$.



## Axioma da União

Dado um conjunto $F$, existe um conjunto cujos elementos são os elementos dos elementos de $F$.
Este conjunto é chamado de **união de** $F$ e é denotado por $$\bigcup_{A\in F}A.$$


#### Exemplo:
Dado o conjunto $$F=\{\{a,b\},\{a,b,c\},\{d\}\},$$ pelo axioma da união, existe o conjunto
$$\bigcup_{A\in F}A=\{a,b,c,d\}.$$

Isso garante a existência do conjunto $$A\cup B=\{x: x\in A \vee y\in B\}$$ a união de $A$ e $B$.

#### Exemplo:
Dado o conjunto $$F=\{\{a,b\},\{a,b,c\},\{d\}\},$$ podemos escrever
$$\bigcup_{A\in F}A=\{a,b\}\bigcup \{a,b,c\}\bigcup \{d\}=\{a,b,c,d\}.$$


#### Exemplo:
Para $S=\{a,b,c,d\}$ e $Q=\{c,d,g,h\}$, temos $S\cup Q=\{a,b,c,d,g,h\}$ e $S\cap Q=\{c,d\}$.


#### Proposição:
Para conjuntos $A$ e $B$, temos:
1. $A\cup A= A$ (idempotência)
1. $A\cup \emptyset = A$
1. $A\cup B=B \cup A$ (comutatividade)
1. $(A\cup B)\cup C=A \cup( B\cup C))$ (transitividade)
1. $(A\subseteq C\wedge B\subseteq C) \Rightarrow A\cup B\subseteq C$
1. $A\subseteq B \Leftrightarrow A\cup B= B$


Podemos também adicionar as propriedades:

#### Proposição:
Para a união e intersecção de conjuntos, temos:

1. $A\cup(B\cap C)=(A\cup B)\cap(A\cup C)$
1. $A\cap(B\cup C)=(A\cap B)\cup(A\cap C)$
1. $A\cup(A\cap B)=A$
1. $A\cap(A\cup B)=A$


#### Demonstração:
Apresentamos a demonstração de que $A\cup(B\cap C)=(A\cup B)\cap(A\cup C)$

$A\cup(B\cap C)$ =  $\{x:x\in A \vee x\in B\cap C\}$
=  $\{x:x\in A \vee (x\in B\wedge x\in C)\}$
=  $\{x:(x\in A \vee x\in B) \wedge (x\in A\vee x\in C)\}$
= $\{x:(x\in A \cup B) \wedge (x\in A\cup C)\}$
= $(A\cup B)\cap(A\cup C)$


Assim, temos $A\cup(A\cap B)=A$ e $A\cap(A\cup B)=A$.


A diferença entre dois conjuntos $A$ e $B$ ou o complementar de $B$ em $A$, $A-B$ ou $A\setminus B$, está definida e é o conjunto $$A\setminus B=\{x:x\in A \wedge x\notin B\}.$$

#### Exemplo:
Dados $A=\{2,5,6\}$, $B=\{3,4,2\}$ e $C=\{1,3,4\}$, temos $A\setminus B = \{5,6\}$, $B\setminus A = \{3,4\}$ e $A\setminus C = \{2,5,6\}$.


#### Exercício:
Mostre que para todo o par de conjuntos $A$ e $B$,
$$A\backslash(A\cap B) = A\backslash B.$$


Entendemos por **diferença simétrica** ou **soma booleana** entre dois conjuntos $A$ e $B$, o conjunto $$A\oplus B=(A\setminus B)\cup(B\setminus A)=(A\cup B)\setminus (A\cap B).$$

As seguintes igualdades são válidas:

$$A\oplus B=B\oplus A,\;\; A\oplus (B\oplus C)=(A\oplus B)\oplus C,\;\;A\oplus \emptyset=A$$

#### Exercício:
Mostre que para todo o conjunto $A$,
$$A\oplus \emptyset = A$$


## Axioma do Conjunto das Partes

Dado um conjunto $A$, existe um conjunto cujos elementos são os subconjuntos de $A$. O conjunto cuja existência é decretada por este axioma chama-se **conjunto das partes** de $A$, ou simplesmente **partes** de $A$, e é representado em notação de conjunto por $\mathcal{P}$, e em notação de construção de conjunto por $$\mathcal{P}(A)=\{x:x\subseteq A\}.$$


#### Exemplo:
Temos:
1. $\mathcal{P}(\emptyset)=\{\emptyset\}$;
1. $\mathcal{P}(\{\emptyset\})=\{\emptyset,\{\emptyset\}\}$;
1. $\mathcal{P}(\{\emptyset,\{\emptyset\}\})=\{\emptyset,\{\emptyset\},\{\{\emptyset\}\},\{\emptyset,\{\emptyset\}\}\}$;
1. $\mathcal{P}(\{0,1\})=\{\emptyset,\{0\},\{1\},\{0,1\}\}$.
1. $\mathcal{P}(\{a,b,c\})=\{\emptyset,\{a\},\{b\},\{c\},\{a,b\},\{b,c\},\{a,c\},\{a,b,c\}\}$




### Compreensão de Listas (List Comprehensions)

A compreensão de listas permite criar uma nova lista aplicando uma expressão a cada item de uma sequência ou iterável.

#### Sintaxe Básica

```python
[nova_expressao for item in iteravel]
```

- `nova_expressao` é a expressão aplicada a cada item (pode ser o próprio item ou alguma transformação dele).
- `item` é o elemento atual da sequência ou iterável.
- `iteravel` é a sequência ou iterável que está sendo percorrida.

#### Exemplo Simples

```python
# Criar uma lista de quadrados de números de 0 a 9
quadrados = [x**2 for x in range(10)]
print(quadrados)  # Saída: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

#### Com Filtro (Condição)

```python
# Criar uma lista de números pares de 0 a 9
pares = [x for x in range(10) if x % 2 == 0]
print(pares)  # Saída: [0, 2, 4, 6, 8]
```

### Compreensão de Conjuntos (Set Comprehensions)

A compreensão de conjuntos segue uma lógica semelhante à de listas, mas cria um conjunto (`set`), que é uma coleção de itens únicos e desordenados.

#### Sintaxe Básica

```python
{nova_expressao for item in iteravel}
```

A principal diferença aqui é o uso de chaves `{}` em vez de colchetes `[]`.



````python
# Criar um conjunto de quadrados únicos de números de 0 a 9
quadrados_unicos = {x**2 for x in range(10)}
print(quadrados_unicos)  # Saída: {0, 1, 64, 36, 4, 9, 16, 49, 81, 25}
````

Com Filtro (Condição)

````python
# Criar um conjunto de números pares de 0 a 9
pares_unicos = {x for x in range(10) if x % 2 == 0}
print(pares_unicos)  # Saída: {0, 8, 2, 4, 6}
````

#### Conjunto das Partes usando List Comprehension

Vamos construir a função para gerar o conjunto das partes de um conjunto utilizando list comprehension e set comprehension.

````python
def conjunto_das_partes(conjunto):
    """
    Dado um conjunto, devolve o conjunto das suas partes.

    Args:
        conjunto (set): O conjunto original.

    Returns:
        set: O conjunto das partes.
    """
    # Começamos com o conjunto vazio
    partes = [set()]

    # Iteramos sobre cada elemento do conjunto original
    for elemento in conjunto:
        # Para cada subconjunto existente, adicionamos o novo elemento para criar novos subconjuntos
        novas_partes = [subconjunto | {elemento} for subconjunto in partes]

        # Adicionamos esses novos subconjuntos ao conjunto das partes
        partes.extend(novas_partes)

    # Convertendo cada subconjunto para frozenset para garantir que sejam imutáveis
    return partes

# Exemplo de uso:
conjunto = {1, 2, 3}
resultado = conjunto_das_partes(conjunto)
print(resultado)
````

### Explicação do Código

1. **Inicialização**:
   - Começamos com uma lista `partes` contendo apenas o conjunto vazio `[set()]`.

2. **Iteração Sobre os Elementos**:
   - Para cada `elemento` no conjunto original, geramos novos subconjuntos adicionando este elemento a todos os subconjuntos existentes (`partes`).

3. **Uso da List Comprehension**:
   - `novas_partes = [subconjunto | {elemento} for subconjunto in partes]` cria uma nova lista de subconjuntos, onde cada subconjunto é o resultado da união (`|`) do subconjunto existente com o conjunto `{elemento}`.

4. **Extensão da Lista de Subconjuntos**:
   - `partes.extend(novas_partes)` adiciona os novos subconjuntos à lista `partes`.

5. **Conversão para Set Comprehension**:
   - `return {frozenset(subconjunto) for subconjunto in partes}` converte cada subconjunto em um `frozenset` e cria um conjunto de `frozensets` para garantir que os subconjuntos sejam únicos e imutáveis.

## Cardinal dum conjunto

Seja $A$ um conjunto. Se o conjunto $A$ tem $n$ elementos, dizemos que $A$ é um \textbf{conjunto finito} e que o \textbf{cardinal} de $A$ é $n$. Neste caso escrevemos \[\sharp A=n \text{ ou } |A|=n.\]
Um conjunto $A$ diz-se \textbf{infinito} se não é finito.

#### Exemplo:
Assim
1. $|\emptyset|=0$;
1. $|\{\emptyset\}|=1$;
1. $|\{\emptyset,\{\emptyset\}\}|=2$.

#### Proposição: [Fórmula de Daniel da Silva]
Para conjuntos finitos $A$ e $B$ tem-se
$$|A\cup B|=|A|+|B|-|A\cap B|.$$


#### Proposição:
Se $A$ é um conjunto finito com $n$ elementos, então $|\mathcal{P}(A)|=2^n$.


Procuremos ilustrar este resultado através da árvore de decisão representada pelo diagrama abaixo.

tree.png


O diagrama tem a seguinte interpretação: Se seleccionar um subconjunto $S$ de $\{a,b,c\}$. O circulo do topo, chamado nó, pode ser interpretado como sendo a pergunta: Está $a$ em $S$? Os dois arcos que saem dele têm como rótulos as duas respostas possíveis (Sim ou Não). Se seguirmos o arco com a resposta apropriada chegamos a um novo nó. Este nó codifica a próxima questão: O $b$ é um elemento de $S$? Seguindo o arco com a resposta certa chegamos a um novo nó. Que codifica a questão: O $c$ é um elemento de $S$? Escolhendo o arco com a resposta certa alcançamos a lista de elementos do conjunto $S$ seleccionado.

Assim, a construção dum subconjunto de $\{a,b,c\}$ corresponde a um percurso no diagrama partido do topo até à base. Existem assim tantos subconjuntos quanto o número de folhas no último nível (a base). Como o número de nós duplica de nível para nível, existem $2^3=8$ nós no último nível (caso a selecção seja feita num conjunto com $n$ elementos teríamos $2^n$ nós na árvore ou seja $2^n$ subconjuntos diferentes).

### Conjuntos das Partes e Cardinalidade

Em contextos mais complexos, como ao calcular o conjunto das partes, entender a cardinalidade é crucial:

```python
def conjunto_das_partes(conjunto):
    partes = [set()]
    for elemento in conjunto:
        novas_partes = [subconjunto | {elemento} for subconjunto in partes]
        partes.extend(novas_partes)
    return {frozenset(subconjunto) for subconjunto in partes}

conjunto = {1, 2, 3}
partes_conjunto = conjunto_das_partes(conjunto)
print("O conjunto das partes é:", partes_conjunto)
print("O cardinal do conjunto das partes é:", len(partes_conjunto))  # Saída: 8
```

Para um conjunto de \( n \) elementos, o conjunto das partes terá \( 2^n \) elementos, o que reflete a explosão combinatorial ao considerar todos os subconjuntos possíveis.


## Par ordenado

Só com os axiomas enunciados pode arquitectar-se
uma teoria de conjuntos satisfatória. Vamos exemplificar um pouco
da teoria com mais algumas definições e teoremas.
Dados dois conjuntos $a$ e $b$, o axioma dos conjuntos elementares garante a existência do
conjunto
$$\{\{a\},\{a,b\}\},$$
a que se chama \textbf{par ordenado} de coordenadas $a$ e $b$ e que
denotamos por $$(a,b).$$
Com base nos axiomas e definições que apresentámos é fácil
demonstrar que
$$(a,b)=(c,d)\Leftrightarrow a=c\wedge b=d.$$
Dados dois conjuntos $A$ e $B$, existe um conjunto cujos
elementos são precisamente todos os pares ordenados $(a,b)$, com $a\in A$ e
$b\in B$, designado de \textit{produto cartsiano} de $A$ por $B$ e que se denota por $$A\times B,$$ i.e. define-se
$$A\times B=\{(a,b):a\in A\wedge b\in B\}.$$
Adoptando-se as seguintes convenções
$$A^2=A\times A,$$
$$A^3=A^2\times A,$$
$$\ldots$$
$$A^n=A^{n-1}\times A.$$
Naturalmente, que se tem:

#### Proposição:
Se $A$ e $B$ são conjuntos finitos com $|A|=n$ e $|B|=m$, então $$|A\times
B|=n\times m.$$


#### Exemplo:
Se $A=\{\alpha,\beta\}$ e $B=\{1,2,3\}$ tem-se

1. $A\times B = \{(\alpha,1),(\alpha,2),(\alpha,3),(\beta,1),(\beta,2),(\beta,3)\}$
1. $B\times A = \{(1,\alpha),(2,\alpha),(3,\alpha),(1,\beta),(2,\beta),(3,\beta)\}$
1. $A\times A = \{(\alpha,\alpha),(\alpha,\beta),(\beta,\alpha),(\beta,\beta)\}$
1. $B\times B = \{(1,1),(1,2),(1,3),(2,1),(2,2),(2,3),(3,1),(3,2),(3,3)\}$

Para três conjuntos $A,B,C$ temos
$$A\times(B\cup C)=(A\times B)\cup (A\times C)$$
$$A\times(B\cap C)=(A\times B)\cap (A\times C)$$

#### Exercício:
Mostre que, para conjuntos $A,B,C$, $A\times(B\cup C)=(A\times B)\cup (A\times C)$


````python
def produto_cartesiano(conjunto1, conjunto2):
    """
    Dado dois conjuntos, devolve o conjunto de todos os tuplos (produto cartesiano)
    onde a primeira componente é elemento do primeiro conjunto e a segunda componente
    é elemento do segundo conjunto.

    Args:
        conjunto1 (set): O primeiro conjunto.
        conjunto2 (set): O segundo conjunto.

    Returns:
        set: O conjunto de tuplos (produto cartesiano).
    """
    return {(x, y) for x in conjunto1 for y in conjunto2}

# Exemplo de uso:
conjunto_a = {1, 2, 3}
conjunto_b = {'a', 'b'}
resultado = produto_cartesiano(conjunto_a, conjunto_b)
print(resultado)
````

#### Definição:
Dois conjuntos $A$ e $B$ são disjuntos se e só se $A\cap B = \emptyset$. Um conjunto de conjuntos é definido por conjuntos disjuntos, se os conjuntos no conjunto são disjuntos dois a dois.


#### Exemplo:
O conjunto de conjuntos $F=\{\{1,3\},\{2,4,5\},\{0,6,7\}\}$, é definida por conjuntos disjuntos dois a dois, já que
$\{1,3\}\cap\{2,4,5\}=\emptyset$, $\{1,3\}\cap\{2,4,5\}=\emptyset$ e $\{2,4,5\}\cap\{0,6,7\}=\emptyset$. Neste caso dizemos que $F$ é uma partição $\{0,1,2,3,4,5,6,7\}$


#### Definição:
 A **diferença entre dois conjuntos** $A$ e $B$ (também designado **o complementar**
  de $B$ em $A$), que se denota por $A-B$ ou $A\setminus B$, está definida e é o conjunto $$A\setminus B=\{x:x\in A \wedge x\notin B\}.$$


Um conjunto $E$ diz-se **universal** se cotem todos os conjuntos em discussão. Neste sentido, para todo o conjunto $A$, $$A\subseteq E$$
ou dito de outra forma, para todo o predicado $P(x)$
$$ E=\{x: P(x)\vee \sim P(x)\}. $$
Por outro lado o conjunto que não cotem elementos é designado de conjunto vazio, ou seja, para todo o predicado $P(x)$ tem-se
$$\emptyset = \{x: P(x)\wedge \sim P(x)\}. $$
Neste sentido, para todo o conjunto $A$,
$$\emptyset\subseteq A \text{ já que } \forall x(x\in \emptyset \rightarrow x\in A)\text{ é uma proposição verdadeira.}$$

#### Definição:
Seja $E$ um conjunto universal. Para todo o conjunto $A$, o complemento de $A$ em $E$, $E-A$ é designado de \textbf{complemento do conjunto} $A$ e é denotado por $\overline{A}$. Neste sentido
$$\overline{A}=\{x\in E:\sim x\in A\}=\{x:x\notin A\}.$$


#### Proposição [Leis de De Morgan]
Para um conjunto universo valem as seguintes igualdades:
1. $\overline{A\cup B}=\overline{A}\cap \overline{B}$
1. $\overline{A\cap B}=\overline{A}\cup \overline{B}$


#### Dem.:
A igualdade $\overline{A\cup B}=\overline{A}\cap \overline{B}$ pode ser demonstrada através

$\overline{A\cup B}$ = $\{x:\sim (x\in A\cup B)\}$
= $\{x:\sim (x\in A \vee x\in B)\}$
= $\{x:\sim (x\in A) \wedge \sim (x\in B)\}$
= $\{x:x\notin A \wedge x\notin B\}$
= $\{x:x\notin A\} \cap \{x:x\notin B\}$
= $\overline{A}\cap \overline{B}$

Assim definido tem-se $$A\setminus B=A\cap \overline{B}.$$

#### Definição:
Entendemos por **diferença simétrica** ou **soma booleana** entre dois conjuntos $A$ e $B$ ao
conjunto $$A\oplus B=(A\setminus B)\cup(B\setminus A)=(A\cup B)\setminus (A\cap B).$$


## Relação Binária

Uma **relação binária** definida de $A$ para $B$ é qualquer subconjunto de $A\times B$.
Uma **relação binária definida em** $A$ é qualquer subconjunto de $A^2=A\times A$.

image.png

Dada uma relação binária $R$:
1. se $(a,b)\in R$ escrevemos $aRb$;
- se $(a,b)\notin R$ escrevemos $aR\!\!\!\!/b$;

Se $R$ é uma relação, existe o conjunto, chamado **domínio** de $R$ e denotado
por $D(R)$, cujos elementos são os conjuntos $x$ para os quais
existe $y$ tal que $(x,y)\in R$, ou seja,
$$D(R)=\{x:\exists y(xRy)\}.$$
De forma semelhante, existe o **conjunto imagem** de $R$ dado por
$$Im(R)=\{y:\exists x(xRy)\}.$$

#### Exemplos:
Qualquer gráfico no plano define uma relação em $\mathbb{R}$. Em particular, o gráfico que descreve o conjunto dos pares reais $(x,y)$ que satisfazem a equação $x^2+y^2=1$ define a relação $$R=\{(x,y):x^2+y^2=1\}$$ em $\mathbb{R}$. Neste caso, $0R1$, $1R0$, $-1R0$, mas $1R\!\!\!\!/1$. Temos também $D(R)=\{x:-1\leq x\leq 1\}$ e $Im(R)=\{y:-1\leq y\leq 1\}$.

  colocando $V$, ou um $F$, na interseção da linha $a$ com a

### Representação de uma Relação numa Matriz em Python

Uma relação binária entre dois conjuntos $A$ e $B$ pode ser representada em uma matriz onde cada elemento da matriz indica se há uma relação entre os elementos correspondentes dos conjuntos $A$ e $B$.

#### Exemplo

Suponha que temos dois conjuntos:
- $A = \{a, b, c\} $
- $B = \{1, 2, 3\} $

E uma relação $R$ definida pelos seguintes pares:
- $(a, 1), (b, 2), (c, 3), (b, 1)$

Podemos representar essa relação em uma matriz $M$, onde:
- As linhas da matriz representam os elementos de $A$ (nesse caso, $ a, b, c $).
- As colunas da matriz representam os elementos de $B$ (nesse caso, $ 1, 2, 3 $).
- Cada célula $M[i][j]$ na matriz será True se o par $(A[i], B[j]) $ estiver na relação $R$, e False caso contrário.

### Definição de Matrizes como Listas de Listas em Python

Em Python, podemos representar matrizes usando listas de listas, onde cada lista interna representa uma linha da matriz.

#### Exemplo de Implementação

Vamos definir a matriz $M$ para representar a relação $R$ descrita anteriormente.



````python
# Conjuntos A e B
conjunto_a = ['a', 'b', 'c']
conjunto_b = [1, 2, 3]

# Relação R
relacao = [('a', 1), ('b', 2), ('c', 3), ('b', 1)]

# Inicialização da matriz com False
matriz = [[False] * len(conjunto_b) for _ in range(len(conjunto_a))]

# Preenchimento da matriz com True onde houver a relação
for (a, b) in relacao:
    indice_a = conjunto_a.index(a)
    indice_b = conjunto_b.index(b)
    matriz[indice_a][indice_b] = True

# Impressão da matriz
for linha in matriz:
    print(linha)
````


### Explicação do Código

1. **Definição dos Conjuntos e da Relação**:
   - `conjunto_a` e `conjunto_b` são listas que representam os conjuntos $A$ e $B$.
   - `relacao` é uma lista de tuplos que descreve os pares na relação $ R$.

2. **Inicialização da Matriz**:
   - Criamos a matriz inicialmente preenchida com `False` usando compreensão de listas.
   ```python
   matriz = [[False] * len(conjunto_b) for _ in range(len(conjunto_a))]
   ```

3. **Preenchimento da Matriz**:
   - Iteramos sobre cada par \( (a, b) \) na lista `relacao`.
   - Para cada par, encontramos os índices correspondentes de \( a \) em `conjunto_a` e de \( b \) em `conjunto_b`.
   - Marcamos a célula correspondente na matriz como `True`.
   ```python
   for (a, b) in relacao:
       indice_a = conjunto_a.index(a)
       indice_b = conjunto_b.index(b)
       matriz[indice_a][indice_b] = True
   ```

4. **Impressão da Matriz**:
   - Finalmente, imprimimos a matriz linha por linha para visualizar a representação da relação $R$ em forma de matriz.

### Exemplo de Saída

Para o exemplo dado:
- `conjunto_a = ['a', 'b', 'c']`
- `conjunto_b = [1, 2, 3]`
- `relacao = [('a', 1), ('b', 2), ('c', 3), ('b', 1)]`

A matriz resultante será:

```
[True,  True, False]
[ True, False,  True]
[False, False,  True]
```

Cada valor `True` na matriz indica que há uma relação entre o elemento correspondente de $A$ e $B$, conforme descrito na lista `relacao`.


````python
def criar_matriz_relacional(conjunto_a, conjunto_b, relacao):
    """
    Cria uma matriz que representa uma relação binária entre dois conjuntos.

    Args:
        conjunto_a (list): Lista dos elementos do conjunto A.
        conjunto_b (list): Lista dos elementos do conjunto B.
        relacao (list of tuples): Lista de tuplos que descreve a relação.
                                  Cada tuplo contém um par (a, b) onde a pertence a A e b pertence a B.

    Returns:
        list of lists: Matriz representando a relação binária.
    """
    # Inicializa a matriz com False
    matriz = [[False] * len(conjunto_b) for _ in range(len(conjunto_a))]

    # Preenche a matriz com True onde houver a relação
    for (a, b) in relacao:
        indice_a = conjunto_a.index(a)
        indice_b = conjunto_b.index(b)
        matriz[indice_a][indice_b] = True

    return matriz

# Exemplo de uso:
conjunto_a = ['a', 'b', 'c']
conjunto_b = [1, 2, 3]
relacao = [('a', 1), ('b', 2), ('c', 3), ('b', 1)]

matriz_relacional = criar_matriz_relacional(conjunto_a, conjunto_b, relacao)
print(matriz_relacional)
````

### Relação complementar
A relação complementar da relação binária $R$ em $A$ é a relação
$$\not R=A^2\backslash R $$
que pode ser definida negando a condição que define $R$.

#### Exemplo:
Seja $R$ a relação binária em $A=\{1,2,3\}$ dada por
$$ aRb\Leftrightarrow a\leq b,\forall a,b\in A.$$

Assim $ a\not R b \Leftrightarrow \sim (a\leq b) \Leftrightarrow a>b,\forall a,b\in A$
ou seja $\not R=\{(2,1),(3,1),(3,2)\}$ que tem por diagrama sagital

que podemos definir pela tabela

| $\not R$ | 1 | 2 | 3 |
|---|---|---|---|
| 1 | V | V | V |
| 2 | F | V | V |
| 3 | F | F | V |

ou

$$
M_{\not R}=
\left[
  \begin{array}{ccc}
    1 & 1 & 1 \\
    0 & 1 & 1 \\
    0 & 0 & 1 \\
  \end{array}
\right]
$$



````python
def complemento_matriz_booleana(matriz):
    """
    Calcula o complementar de uma matriz booleana.

    Args:
        matriz (list of lists): Matriz booleana representada como lista de listas.

    Returns:
        list of lists: Matriz complementar booleana.
    """
    complemento = []
    for linha in matriz:
        linha_complementar = [not valor for valor in linha]
        complemento.append(linha_complementar)
    return complemento

# Exemplo de uso:
matriz_booleana = [
    [True, False, True],
    [False, True, False],
    [True, False, True]
]

complemento = complemento_matriz_booleana(matriz_booleana)
for linha in complemento:
    print(linha)
````

## Reflexividade

Uma relação binária $R$ definida no conjunto $A$ diz-se:

1. **Reflexiva** se: $\forall x\in A: xRx$.
1. **Não reflexiva** se: $\exists x\in A: x\not R x$.
1. **Anti-reflexiva** se: $\forall x,y\in A: xRy\rightarrow x\neq y$.



#### Exemplos:

Note que, uma relação finita é reflexiva se no diagrama sagital cada ponto tem um laço, ou caso na matriz da relação todos os elementos da diagonal são 1.

#### Exemplos:

Exemplos de relações anti-reflexivas no conjunto $A=\{1,2,3\}$

image.png




````python
def is_reflexive(relacao):
    """
    Verifica se uma relação binária é reflexiva.

    Args:
        relacao (set): Conjunto de tuplos que descrevem a relação.

    Returns:
        bool: True se a relação for reflexiva, False caso contrário.
    """
    # Obter todos os elementos envolvidos na relação
    elementos = {x for tuplo in relacao for x in tuplo}

    # Verificar se para cada elemento (x, x) está presente na relação
    for x in elementos:
        if (x, x) not in relacao:
            return False
    return True

# Exemplos de uso:
relacao_reflexiva = {(1, 1), (2, 2), (3, 3), (1, 2), (2, 1)}
relacao_nao_reflexiva = {(1, 2), (2, 3), (3, 1)}

print(is_reflexive(relacao_reflexiva))  # Saída: True
print(is_reflexive(relacao_nao_reflexiva))  # Saída: False
````

### Simetria

 Uma relação binária $R$ definida no conjunto $A$ diz-se:

1. **Simétrica** se: $\forall x,y\in A: xRy\rightarrow yRx$.
1. **Não simétrica** se: $\exists x,y\in A: xRy \wedge y\not R x$.
1. **Anti-simétrica** se: $\forall x,y\in A: xRy \wedge yRx \rightarrow x = y$.


#### Exemplos

Uma matriz $M_R$ representa uma relação simétrica se é simétrica.

#### Exemplos
Exemplos de relações anti-simétricas no conjunto $A=\{1,2,3\}$

````python
def is_symmetric(relacao):
    """
    Verifica se uma relação binária é simétrica.

    Args:
        relacao (set): Conjunto de tuplos que descrevem a relação.

    Returns:
        bool: True se a relação for simétrica, False caso contrário.
    """
    for (a, b) in relacao:
        if (b, a) not in relacao:
            return False
    return True

# Exemplo de uso:
relacao_simetrica = {(1, 2), (2, 1), (3, 3), (4, 5), (5, 4)}
relacao_nao_simetrica = {(1, 2), (2, 3), (3, 1)}

print(is_symmetric(relacao_simetrica))  # Saída: True
print(is_symmetric(relacao_nao_simetrica))  # Saída: False
````

## Transitividade

Uma relação binária $R$ definida no conjunto $A$ diz-se:

1. **Transitiva** se: $\forall x,y,z\in A: xRy \wedge yRz \rightarrow xRz$.
1. **Não transitiva** se: $\exists x,y,z\in A: xRy \wedge yRz \wedge x\not R z$.

### Exemplos

Exemplos de relações transitivas no conjunto $A=\{1,2,3,4\}$

image.png


### Exemplos

Exemplo duma relação não transitiva no conjunto $A=\{1,2,3,4\}$

image.png

````python
def is_transitiva(relacao):
    """
    Verifica se uma relação binária é transitiva.

    Args:
        relacao (set): Conjunto de tuplos que descrevem a relação.

    Returns:
        bool: True se a relação for transitiva, False caso contrário.
    """
    for (a, b) in relacao:
        for (c, d) in relacao:
            if b == c and (a, d) not in relacao:
                return False
    return True

# Exemplo de uso:
relacao_transitiva = {(1, 2), (2, 3), (1, 3)}
relacao_nao_transitiva = {(1, 2), (2, 3), (1, 4)}

print(is_transitiva(relacao_transitiva))  # Saída: True
print(is_transitiva(relacao_nao_transitiva))  # Saída: False
````



### Explicação do Código

1. **Definição da Função `is_transitiva`**:
   - Recebe um conjunto `relacao` que representa a relação binária.

2. **Verificação da Transitividade**:
   - Iteramos sobre cada par $ (a, b) $ na relação.
   - Para cada par, verificamos todos os pares $ (c, d) $ na relação.
   - Verificamos se há um elemento intermediário $ b $ que conecta $ a $ a $ d $através de $ c $:
     ```python
     if b == c and (a, d) not in relacao:
         return False
     ```
     Se encontrarmos um par $(a, b) $ e $ (b, c) $ na relação, mas não encontrarmos $ (a, c) $, então a relação não é transitiva e retornamos `False`.

3. **Retorno da Transitividade**:
   - Se todos os pares na relação passarem na verificação transitiva, a função retorna `True`.

4. **Exemplos de Uso**:
   - Testamos a função com exemplos de relações transitivas e não transitivas para validar seu comportamento.


### Fecho
Dada uma relação binária $R$ definida em $A$ definimos:
  1. O **fecho transitivo** da relação $R$ é a relação $R^+$ de $A$ dada por:
  
    a. se $aRb$ então $aR^+b$,

    b. se $aR^+b$ e $bR^+c$ então $aR^+c$.

  2. O **fecho reflexivo** da relação $R$ é a relação $R^-$ de $A$ dada por:
  $$
  R^-=R\cup \{(a,a)|a \in A\}.
  $$
  3. O **fecho transitivo e reflexivo** da relação $R$ é a relação $R^*$ de $A$ dada por:
  $$
  R^*=R^+\cup R^-.
  $$

#### Exemplo

Exemplificação do fecho transitivo duma relação $R=\{(1,2),(2,3),(3,4)\}$ em $A=\{1,2,3,4\}$

image.png

#### **Proposição**
Uma relação binária $R$ é transitiva se e só se $R=R^+$ i.e. $R$ for igual ao seu fecho transitivo.



````python
def fecho_transitivo(relacao, conjunto):
    """
    Retorna o fecho transitivo de uma relação binária definida em extensão sobre um conjunto.

    Args:
        relacao (set): Conjunto de tuplos que descrevem a relação binária.
        conjunto (set): Conjunto sobre o qual a relação está definida.

    Returns:
        set: Fecho transitivo da relação binária.
    """
    fecho = set(relacao)  # Inicializa o fecho com a relação original

    while True:
        novo_adicionado = False
        for (a, b) in fecho.copy():  # Utiliza uma cópia para evitar problemas durante a iteração
            for (c, d) in fecho.copy():
                if b == c and (a, d) not in fecho:
                    fecho.add((a, d))
                    novo_adicionado = True
        if not novo_adicionado:
            break

    return fecho

# Exemplo de uso:
relacao = {(1, 2), (2, 3), (3, 4)}
conjunto = {1, 2, 3, 4}

fecho = fecho_transitivo(relacao, conjunto)
print("Fecho transitivo da relação:", fecho)
````

### Explicação do Código

1. **Definição da Função `fecho_transitivo`**:
   - Recebe dois argumentos: `relacao` (conjunto de tuplos que descrevem a relação binária) e `conjunto` (conjunto sobre o qual a relação está definida).

2. **Inicialização do Fecho**:
   - Inicializa o fecho com o conjunto `relacao`, pois o fecho transitivo começa com as relações existentes.

3. **Iteração e Adição de Novas Relações**:
   - Utiliza um loop `while True` que continua até não serem adicionadas novas relações.
   - Itera sobre cada par $ (a, b)$ e $ (c, d) $ na cópia atual do fecho.
   - Se encontrar um elemento intermediário $ b $ que conecta $ a $ a $ d $ através de $ c $ (ou seja, $ (a, b) $ e $ (b, d) $ estão no fecho mas $(a, d) $ não está), adiciona $ (a, d) $ ao fecho.

4. **Condição de Parada**:
   - O loop continua até que nenhuma nova relação seja adicionada na iteração, garantindo que o fecho transitivo esteja completo.

5. **Retorno do Fecho Transitivo**:
   - Retorna o fecho transitivo completo da relação binária.


Notas: O uso de fecho.copy() é uma prática recomendada sempre que se precisa iterar e modificar uma estrutura de dados ao mesmo tempo, garantindo a consistência dos dados e a correta execução dos algoritmos. Ele é especialmente útil em Python e em outras linguagens de programação que possuem gestão explícito de memória e referências, ajudando a evitar armadilhas comuns relacionadas à modificação de estruturas de dados durante a iteração.

### Utilização de `fecho.copy()` na Implementação

Na função `fecho_transitivo` que calcula o fecho transitivo de uma relação binária, utilizamos `fecho.copy()` por algumas razões importantes:

1. **Evitar Modificações Durante a Iteração**:
   - Ao iterar sobre um conjunto e modificá-lo ao mesmo tempo (adicionando novos elementos), pode ocorrer um comportamento inesperado ou um erro de tempo de execução, dependendo da linguagem de programação ou do ambiente.
   - `fecho.copy()` cria uma cópia independente do conjunto `fecho`, permitindo que a iteração continue sem interferências das modificações que estamos realizando.

   - A cópia do conjunto garante que as operações de iteração e modificação não afetem a estrutura original do conjunto `fecho` até que a iteração esteja completa.

### Importância na Implementação de Algoritmos

- **Segurança e Integridade dos Dados**:
  - Em algoritmos que envolvem iteração e modificação de estruturas de dados complexas, como conjuntos, listas ou dicionários, o uso de `copy()` ou equivalentes é crucial para evitar problemas de referência ou de sincronização de dados.

- **Performance e Eficiência**:
  - Embora criar uma cópia adicional possa consumir um pouco mais de memória e tempo de execução, em muitos casos a segurança e a garantia de que o algoritmo funcione corretamente compensam esse custo adicional.
  - Além disso, em Python, operações com conjuntos geralmente são eficientes, o que minimiza o impacto do uso de `copy()`.


Para verificar se uma relação binária $ R $ é transitiva podemos comparar a relação $ R $ com o seu fecho transitivo obtido usando a função que implementamos anteriormente.


````python
def verificar_transitividade(relacao, conjunto):
    """
    Verifica se uma relação binária é transitiva.

    Args:
        relacao (set): Conjunto de tuplos que descrevem a relação binária.
        conjunto (set): Conjunto sobre o qual a relação está definida.

    Returns:
        bool: True se a relação for transitiva, False caso contrário.
    """
    fecho = fecho_transitivo(relacao, conjunto)
    return fecho == relacao

# Exemplo de uso:
relacao = {(1, 2), (2, 3), (1, 3)}
conjunto = {1, 2, 3}

transitiva = verificar_transitividade(relacao, conjunto)
print("A relação é transitiva?", transitiva)
````
